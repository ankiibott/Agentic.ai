```python
import os
from dotenv import load_dotenv
from google import genai

load_dotenv()

# Instantiating the client directly handles the new AQ. key style seamlessly
client = genai.Client()

blood_report = "Hemoglobin: 14.2 g/dL (Normal: 13.5-17.5)" # Dummy text for verification

extraction_prompt = f"""
From the blood report below, extract ALL test values and classify each one as HIGH, LOW, or NORMAL.
Blood Report:
{blood_report}
"""

print("Sending native request directly to Google...")
try:
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=extraction_prompt,
    )
    print("=== SUCCESS ===")
    print(response.text)
except Exception as e:
    print("=== STILL FAILING NATIVELY ===")
    print(e)
```
